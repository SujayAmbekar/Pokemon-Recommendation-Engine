```python
# import statements
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np

# Reading CSV file
PokeData = pd.read_csv('/Users/sujayambekar/Desktop/PokemonData.csv')

# List created using user inputed pokemons
Pokemons = []
                        ##'E:/College/PESIO_RecommendationEngine/PokemonData.csv'
for i in range(6):
    Poke = input("Enter your Pokemon: ")
    Pokemons.append(Poke)

# List of different columns of the CSV files
Names = list(PokeData['Name'])
HPLS = list(PokeData['HP'])
AttackLS = list(PokeData['Attack'])
DefenseLS = list(PokeData['Defense'])
SpeedLS = list(PokeData['Speed'])
SpAttackLS = list(PokeData['SpAtk'])
SpDefenseLS = list(PokeData['SpDef'])

# Creating lists of user's each Pokemon's data
HPPK = []
AttackPK = []
DefensePK = []
SpeedPK = []
SpAttackPK = []
SpDefensePK = []
for i in Pokemons:
    for j in Names:
        if i == j:
            k = Names.index(j)
            HPPK.append(HPLS[k])
            AttackPK.append(AttackLS[k])
            DefensePK.append(DefenseLS[k])
            SpeedPK.append(SpeedLS[k])
            SpAttackPK.append(SpAttackLS[k])
            SpDefensePK.append(SpDefenseLS[k])

# Creating vectors out of the User inputed Pokemon Data.
Poke1 = np.array([AttackPK[0], DefensePK[0], SpeedPK[0], HPPK[0], SpAttackPK[0], SpDefensePK[0]]).reshape(1, 6)
Poke2 = np.array([AttackPK[1], DefensePK[1], SpeedPK[1], HPPK[1], SpAttackPK[1], SpDefensePK[1]]).reshape(1, 6)
Poke3 = np.array([AttackPK[2], DefensePK[2], SpeedPK[2], HPPK[2], SpAttackPK[2], SpDefensePK[2]]).reshape(1, 6)
Poke4 = np.array([AttackPK[3], DefensePK[3], SpeedPK[3], HPPK[3], SpAttackPK[3], SpDefensePK[3]]).reshape(1, 6)
Poke5 = np.array([AttackPK[4], DefensePK[4], SpeedPK[4], HPPK[4], SpAttackPK[4], SpDefensePK[4]]).reshape(1, 6)
Poke6 = np.array([AttackPK[5], DefensePK[5], SpeedPK[5], HPPK[5], SpAttackPK[5], SpDefensePK[5]]).reshape(1, 6)
Pokes = [Poke1, Poke2, Poke3, Poke4, Poke5, Poke6]

# Creating dictionaries for each of the 6 pokemon's cosine similarities with the rest of the Pokemons in the Data
RecomPokes = []
for j in Pokes:
    CosDict = {str(i): cosine_similarity(j,
                                         np.array([AttackLS[Names.index(i)], DefenseLS[Names.index(i)],
                                                   SpeedLS[Names.index(i)], HPLS[Names.index(i)],
                                                   SpAttackLS[Names.index(i)],
                                                   SpDefenseLS[Names.index(i)]]).reshape(1, 6))
               for i in Names}
# Dictionaries converted into lists and sorted in descending order of their cosine similarities
    j = sorted(CosDict.items(), key=lambda x: -x[1])
    for k in range(1, 5):
        RecomPokes.append(j[k][0])

# Creating a list of uniques recommended Pokemons from above created lists.
RecomPokes = list(set(RecomPokes))
for i in RecomPokes:
    if i in Pokemons:
        RecomPokes.remove(i)
print("You should try catching these Pokemons:")
for i in range(len(RecomPokes)):
    print(RecomPokes[i])
```
